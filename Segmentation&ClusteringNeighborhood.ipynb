```python
import requests
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = requests.get(url).content
df_list = pd.read_html(html)
df = df_list[0]
print(df)
df.to_csv('postalcodescanada.csv')    #Import data and store the resulting data in data frame
```

```python
df.head()
```

```python
#Replacing 'Not assigned' value in 'Borough' column with null value
df1=df
df1['Borough'] = df1['Borough'].replace({'Not assigned':None})
df1.head()
```

```python
#Drop null values in 'Borough' column
df1.dropna(axis=0, inplace=True)
df1.head(10)
```

```python
# Re index the data frame after dropping null values from Borough column
df1 = df1.reset_index()
del df1['index']
df1.head(10)
```

```python
# To replace 'Not available' in Neighbourhood column to none
df1['Neighbourhood'] = df1['Neighbourhood'].replace({'Not assigned':None})
df1.head(10)
```

```python
#Replace None values with corresponding row values of Borough
df1.Neighbourhood.fillna(df1.Borough, inplace=True)
df1.head(10)
```

```python
#Merge multiple postcode values under single post code
df1 = df1.groupby('Postcode').agg({'Borough':'first', 
                             'Neighbourhood': ', '.join}).reset_index()
df1.head(10)
```
